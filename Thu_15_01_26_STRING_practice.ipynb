# **String Function**

find(), index(), isalnum(), isdigit(), isnumeric(), islower(), isupper(), len(), lower(), upper(), replace(), join(), split()

```python
sfd = "python programming"
print(sfd.find("pro"))
```

```python
si = "python programming"
print(si.index("pro"))
```

```python
sal = "Python123"
print(sal.isalnum())
```

```python
sd = "12345"
print(sd.isdigit())
```

```python
snu = "268376528"
print(snu.isnumeric())
```

```python
silow = "python"
print(silow.islower())
```

```python
siu = "PYTHON"
print(siu.isupper())
```

```python
slen = "Python"
print(len(slen))
```

```python
sl = "PYTHON"
print(sl.lower())
```

```python
us = "python"
print(us.upper())
```

```python
sr = "I like Java"
print(sr.replace("Java", "Python"))
```

```python
words = ["Python", "is", "easy"]
print(" ".join(words))
```

```python
ss = "Python is easy"
print(ss.split())
```
