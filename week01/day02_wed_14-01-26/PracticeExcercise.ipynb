1. Given the string “PYTHONPROGRAMMING”, extract every third character starting from the second character.

```python
strpy = "PYTHONPROGRAMMING"

print(strpy)

print("with idx no  =  ", strpy[3:4],strpy[6:7],strpy[9:10],strpy[12:13],strpy[15:16])

print("Without idx no =  ",strpy[2:3],strpy[5:6],strpy[8:9],strpy[11:12],strpy[14:15])
```

2. From the string “DEVELOPER”, reverse the string except for the first and last characters.

```python
dev="DEVELOPER"
print(dev)
print(dev[0]+dev[7]+dev[6]+dev[5]+dev[4]+dev[3]+dev[2]+dev[1]+dev[-1])
#OR
print(dev[0:1]+dev[-2:-1]+dev[6:7]+dev[5:6]+dev[4:5]+dev[3:4]+dev[2:3]+dev[1:2]+dev[-1:])
```

3. Using the string “DATA_SCIENCE_AND_AI”, obtain the characters starting from the twelfth position up to the sixth character from the end, in reverse order.

```python
ds='DATA_SCIENCE_AND_AI'
print(ds[-7:-6])
#OR
print(ds[12:13])
```

4. For the string “ABCDEFGHIJKLMNO”, extract the characters that are at even positions from the second half of the string.

```python
s='ABCDEFGHIJKLMNO'
print(s)

print(s[8],s[10],s[12],s[14])
```

5. Using the string “COMPUTERENGINEERING”, rotate the string to the left by four characters.

```python
cs='COMPUTERENGINEERING'
print(cs[4:]+cs[0:4])
```

6. From the string “PYTHON”, perform a slicing operation that results in an empty string

7. Using the string “0123456789”, extract characters starting from the second character from the end up to the tenth character from the end, moving backward with a step of two.

```python
n='0123456789'
print(n[-2:-1],n[-4:-3],n[-6:-5],n[-8:-7],n[-10:-9])
```
