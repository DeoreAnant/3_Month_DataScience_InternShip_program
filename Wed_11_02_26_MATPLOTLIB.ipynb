```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=[1,3,4,5,7,6,9]
y=[10,20,50,25,60,50,90
   ]
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='black')
plt.title('line graph')
plt.xlabel('Y-axis')
plt.ylabel('X-axis')
plt.show()
```

```python
plt.plot(x,y,color='orange',linestyle='dotted',marker='^',linewidth=1.6)
plt.title('line graph')
plt.xlabel('y axis')
plt.ylabel('x axis')
plt.show()
```

```python
plt.bar(x,y,color='red')
plt.title("bar graph")
```

```python
plt.hist(x,bins=6)
plt.title("histogram")
plt.show()
```

```python
plt.scatter(x,y,color='black')
plt.legend()
plt.title("scatter plot")
plt.show()
```

```python
plt.pie(x,labels=y)
plt.title("pie chart")
plt.show()
```
