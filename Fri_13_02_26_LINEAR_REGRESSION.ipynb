# **1. Linear Regression**

Create Linear Regression model on given dataset

### **1. Laptop price dataset**

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/content/laptop_price.csv')
df.head()
```

```python
df.columns
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df
```

```python
df.head()
```

```python
plt.scatter(df['RAM'],df['Price'], marker='+', color='red')
plt.plot(df['RAM'],df['Price'],color='lightgreen')
plt.xlabel('RAM')
plt.ylabel('Price')
plt.title('RAM  Vs  Price')
plt.show()
```

```python
X = df[['RAM']]
Y = df[['Price']]
```

```python
linear_regression_model = LinearRegression()
linear_regression_model.fit(X,Y)
```

```python
linear_regression_model.predict([[128]])
```

```python
linear_regression_model.predict([[64]])
```

```python
linear_regression_model.predict([[30]])
```

```python
linear_regression_model.predict([[32]])
```

```python
plt.scatter(df['RAM'],df['Price'], marker='+')
plt.plot(X,linear_regression_model.predict(X), color='red')
plt.xlabel('RAM')
plt.ylabel('Price')
plt.title('RAM  Vs  Price')
plt.show()
```
