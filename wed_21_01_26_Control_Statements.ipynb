# **Control Statements**

```python
n = int(input("Enter a number n = "))
b =int(input("Enter a base b = "))

if n>b:
  print("n is greater than b")
elif n==b:
  print("n is equal to b")
else:
  print("n is less than b")
```

```python
#max number in 3
x = int(input("Enter a number x = "))
y = int(input("Enter a number y = "))
z = int(input("Enter a number z = "))

if x>y and x>z:
  print("x is greater = ",x)
elif y>z and y>x:
    print("y is greater = ", y)
else:
    print("z is greater = ", z)
```

```python
#program of all type of function return not return
def add(a,b):
  return a+b
def sub(a,b):
  return a-b
def mul(a,b):
  return a*b
def div(a,b):
  return a/b

a = int(input("Enter a = "))
b = int(input("Enter b = "))
o = input("Enter + - * / ")
if o == '+':
  print(add(a,b))
  print(add(a,b))
elif o == '-':
  print(sub(a,b))
elif o == '*':
  print(mul(a,b))
elif o == '/':
  print(div(a,b))
```
